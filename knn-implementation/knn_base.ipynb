# Classificador k-NN

Implemente a seguinte classe em python que implementa um classificador k-NN.
Observações:
1. Só é necessário implementar as funções fit e predict;
2. Pode-se adicionar atributos à classe conforme seja necessário;
3. O restante das células apenas aplicam a classe definida na base Iris e não deve ser modificado.

```python
import math

class KNN_Classifier:
    neighbors = []
    
    def getDistance(self, from_point, to_point):
        distance = 0
        for axis in range(len(from_point)):
            distance += pow((from_point[axis] - to_point[axis]), 2)
        return math.sqrt(distance)

    def fit (self, X_train, y_train):
        for i in range(len(X_train)):
            neighbor = {"label": y_train[i], "values": X_train[i]}
            self.neighbors.append(neighbor)

        print('knn fit')
    
    def predict(self, X_test, k):
        y = []
        for test in X_test: 
            best_neighbors = []
            # Define the first k's as best_neighbors to init 
            for i in range(k):
                best_neighbors.append({"neighbor": self.neighbors[i], "distance": self.getDistance(test, self.neighbors[i]["values"])})

            for i in range(len(self.neighbors)):
                distance = self.getDistance(test, self.neighbors[i]["values"])
                for d in range(k):
                    if(distance < best_neighbors[d]["distance"]):
                        best_neighbors[d] = {"neighbor": self.neighbors[i], "distance": distance}
                        break
                print("the best neighors of", test, " is ", best_neighbors)


            # Check the label tha most appears in best_neighbors
            best_label = {"i_label_count": 0}
            for nc in best_neighbors:
                i_label_count = 0
                for nt in best_neighbors:
                    if nc["neighbor"]["label"] == nt["neighbor"]["label"]:
                         i_label_count += 1
                if(i_label_count > best_label["i_label_count"]):
                    best_label["i_label_count"] = i_label_count
                    best_label["label"] = nc["neighbor"]["label"]  

            # Add to y this label as predict value
            y.append(best_label["label"])

        return y
```

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data # use only 'sepal length and sepal width'
y = iris.target
```

### Divide a base de dados utilizando para treinamento e teste.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Cria o classificador utilizando a classe definida acima
Treina utilizando X_train e y_train e faz a predição em X_test.

```python
knn = KNN_Classifier()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test, 5)
```

### Avalia o modelo utilizando matriz confusão e as métricas conhecidas (precision, recall, etc)

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
